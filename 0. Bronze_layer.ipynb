```python
Ingestion_items = [{
    "src": "crm app",
    "path": "/Volumes/workspace/0_bronze/src_systems/crm/cust_data*.csv",
    "src_fileformat":"csv",
    "tablename": "workspace.0_bronze.cust_data_tbl"
    
}]
```

```python
for item in Ingestion_items:
    if item["src_fileformat"]=="csv":
        print(item["src"],"-->",item["path"],"-->",item["tablename"])
        df =(spark.read.option("header", "true")
                        .option("inferSchema", "true") 
                        .csv(item["path"])
        )
        df=df.withColumn("ingest_ts", current_timestamp() )
        display(df)
        df.write.mode("append").format("delta").saveAsTable(item["tablename"])
```
